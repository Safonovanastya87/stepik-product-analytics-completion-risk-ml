## USER x STEP level

```r
library(data.table)

# 1. Базовые метрики активности
user_metrics <- events_data[, .(
  n_discovered = sum(action == "discovered"),
  n_viewed = sum(action == "viewed"),
  n_started_attempt = sum(action == "started_attempt"),
  n_passed = sum(action == "passed"),
  first_timestamp = min(timestamp_dt)
), by = user_id]

# 2. Доли правильных и неправильных сабмишнов
sub_status_ratios <- submissions_data[, .(
  correct_ratio_sub = if (.N > 0) sum(submission_status == "correct") / .N else NA_real_,
  wrong_ratio_sub   = if (.N > 0) sum(submission_status == "wrong") / .N else NA_real_
), by = user_id]

# 3. Dropoff (только задания)
started <- events_data[action=="started_attempt", .(user_id, step_id)]
passed  <- events_data[action=="passed", .(user_id, step_id)]
setnames(passed, "step_id", "step_id_pass")

dropoff <- merge(started, passed, by="user_id", all.x=TRUE, allow.cartesian=TRUE)
dropoff[, drop_flag := is.na(step_id_pass)]

dropoff_user <- dropoff[, .(
  dropoff_steps = sum(drop_flag),
  first_dropoff_step = step_id[which(drop_flag)[1]]
), by = user_id]

# 4. Последняя активность пользователя (события и сабмишны)
last_event_ts <- events_data[, .(last_event_ts = max(timestamp_dt)), by=user_id]
last_submission_ts <- submissions_data[, .(last_submission_ts = max(timestamp_dt)), by=user_id]

last_ts <- merge(last_event_ts, last_submission_ts, by="user_id", all=TRUE)
last_ts[, last_timestamp := pmax(last_event_ts, last_submission_ts, na.rm=TRUE)]
last_ts <- last_ts[, .(user_id, last_timestamp)]

# 5. Объединяем все метрики
user_metrics <- merge(user_metrics, sub_status_ratios, by="user_id", all.x=TRUE)
user_metrics <- merge(user_metrics, dropoff_user, by="user_id", all.x=TRUE)
user_metrics <- merge(user_metrics, last_ts, by="user_id", all.x=TRUE)

# 6. Продолжительность курса и частота активности
user_metrics[, course_duration := as.numeric(difftime(last_timestamp, first_timestamp, units="days"))]
user_metrics[, activity_frequency := ifelse(course_duration > 0,
                                            (n_discovered + n_viewed + n_started_attempt + n_passed)/course_duration,
                                            NA_real_)]

# 7. Оставляем только нужные колонки
user_metrics <- user_metrics[, .(
  user_id,
  n_discovered, n_viewed, n_started_attempt, n_passed,
  correct_ratio_sub, wrong_ratio_sub,
  dropoff_steps, first_dropoff_step,
  course_duration, activity_frequency
)]
    
    
glimpse(user_metrics)
```

```r
actions_diagnostic <- rbindlist(lapply(c(
  "n_discovered",
  "n_viewed",
  "n_started_attempt",
  "n_passed"
), function(m) {
  x <- user_metrics[[m]]
  data.table(
    metric      = m,
    min         = min(x, na.rm = TRUE),
    p25         = quantile(x, 0.25, na.rm = TRUE),
    p50         = quantile(x, 0.50, na.rm = TRUE),
    p75         = quantile(x, 0.75, na.rm = TRUE),
    mean        = mean(x, na.rm = TRUE),
    p90         = quantile(x, 0.9,  na.rm = TRUE),
    p99         = quantile(x, 0.99, na.rm = TRUE),
    max         = max(x, na.rm = TRUE),
    pct_zero    = mean(x == 0, na.rm = TRUE),
    pct_gt_198  = mean(x > 198, na.rm = TRUE)  
  )
}))

actions_diagnostic
```

```r
# Рассчитываем 95%-перцентиль для нужных метрик
p95_viewed <- quantile(user_metrics$n_viewed, 0.95, na.rm = TRUE)
p95_started <- quantile(user_metrics$n_started_attempt, 0.95, na.rm = TRUE)

# Вектор пользователей с экстремальными значениями (верхние 5%)
extreme_user_ids <- user_metrics %>%
  filter(
    n_viewed > p95_viewed |
    n_started_attempt > p95_started
  ) %>%
  pull(user_id)

# Фильтруем их из датафрейма и считаем статистики без верхних 5%
actions_diagnostic_no95p <- rbindlist(lapply(c(
  "n_discovered",
  "n_viewed",
  "n_started_attempt",
  "n_passed"
), function(m) {
  x <- user_metrics[!user_id %in% extreme_user_ids, ..m][[1]]
  
  data.table(
    metric      = m,
    min         = min(x, na.rm = TRUE),
    p25         = quantile(x, 0.25, na.rm = TRUE),
    p50         = quantile(x, 0.50, na.rm = TRUE),
    p75         = quantile(x, 0.75, na.rm = TRUE),
    mean        = mean(x, na.rm = TRUE),
    p90         = quantile(x, 0.9,  na.rm = TRUE),
    p99         = quantile(x, 0.99, na.rm = TRUE),
    max         = max(x, na.rm = TRUE),
    pct_zero    = mean(x == 0, na.rm = TRUE),
    pct_gt_198  = mean(x > 198, na.rm = TRUE)
  )
}))

cat("Метрики активности пользователей с 95%-перцентилем\n")
actions_diagnostic
cat("Метрики активности пользователей без 95%-перцентиля\n")
actions_diagnostic_no95p
```

```r
#Сравнение метрик активности пользователей с верхними 5% и без них:
#1. Верхние 5% самых активных пользователей (экстремалы) сильно искажают средние и верхние перцентили, особенно для n_viewed и n_started_attempt.
#2. Исключение этих экстремалов практически не меняет медианы и нижние квартели, что подтверждает стабильность поведения основной массы пользователей.
#3. Верхние хвосты (p90, p99, max) значительно снижаются, делая статистику более репрезентативной для типичного пользователя.
#4. Доля нулевых значений (pct_zero) и доля пользователей с аномально высокой активностью (pct_gt_198) почти не меняются, что показывает, что массовое поведение остаётся неизменным.
#5. Вывод: метрики без верхних 5% дают более чистое и стабильное представление о поведении основной аудитории, в то время как полные метрики полезны для выявления суперпользователей и аномальных паттернов.
```

```r
# Выбираем метрики
metrics <- c("n_discovered",
  "n_viewed",
  "n_started_attempt",
  "n_passed")

# Подготавливаем данные
plot_data <- rbindlist(lapply(metrics, function(m) {
  data.table(
    metric = m,
    value = user_metrics[[m]],
    type = "All users"
  )
}))

plot_data_no95 <- rbindlist(lapply(metrics, function(m) {
  p95 <- quantile(user_metrics[[m]], 0.95, na.rm = TRUE)
  x <- user_metrics[[m]][user_metrics[[m]] <= p95]
  data.table(
    metric = m,
    value = x,
    type = "Without top 5%"
  )
}))

plot_data <- rbind(plot_data, plot_data_no95)

# Создаем график
ggplot(plot_data, aes(x = metric, y = value, fill = type)) +
  geom_boxplot(outlier.shape = NA, alpha = 0.5, position = position_dodge(width = 0.8)) +
  geom_violin(aes(color = type), alpha = 0.2, position = position_dodge(width = 0.8), scale = "width") +
  scale_y_continuous(trans = "log10") +  # логарифмическая шкала для наглядности
  labs(
    title = "Сравнение распределения метрик активности пользователей",
    x = "Метрика",
    y = "Значение (лог10)",
    fill = "",
    color = ""
  ) +
  theme_minimal(base_size = 14) +
  theme(legend.position = "top")
```

```r
# Выводы по распределениям метрик активности пользователей:
# 1. Все основные метрики (n_discovered, n_viewed, n_started_attempt, n_passed)
#    имеют правостороннюю асимметрию, что типично для пользовательского поведения:
#    большинство пользователей малo- и среднеактивны, небольшая доля — сверхактивна.
# 2. Наиболее выраженный правый хвост наблюдается у n_viewed — метрика чувствительна
#    к повторным просмотрам и «скроллинговому» поведению.
# 3. Распределение n_started_attempt также правостороннее, но асимметрия заметно слабее,
#    так как начало попытки требует осознанного действия пользователя.
# 4. Исключение верхних 5% самых активных пользователей существенно сокращает хвосты
#    и снижает влияние выбросов на средние и верхние перцентили.
# 5. При этом медианы и межквартильный размах почти не меняются, что подтверждает
#    устойчивость и репрезентативность поведения основной массы пользователей.
# 6. Вывод: для продуктовой диагностики и сравнения пользователей целесообразно
#    использовать метрики без верхних 5%, сохраняя полное распределение отдельно
#    для анализа суперпользователей и аномального поведения.
```

```r
steps <- sort(unique(events_data$step_id))
step_score_vec <- rep(0, length(steps))

score <- list(
  `1` = c(3,14,15,18,19,22,30,31,32,33,38,39,40,47,49,53,54,55,56,57,
          62,64,65,68,73,74,76,80,81,82,83,95,100,104,108,109,111,
          122,123,131,133,138,139,140,141,147,148,149,155,156,157,
          161,162,163,169,170,172,177,185,186),

  `2` = c(41,63,94,99,113,114,115,125,132,179,189),

  `3` = c(126,180)
)

# Векторизация загрузки score
step_score_vec[unlist(score)] <- rep(names(score), lengths(score))

steps_table <- data.table(
  step_id = steps,
  step_score = step_score_vec
)

steps_table[, is_assignment := step_id %in% submissions_data$step_id]

## было сравнение с ноутбуком с ML, присвоение скор прошло корректно
```

```r
step_discovered <- events_data %>%
  filter(action == "discovered") %>%
  distinct(step_id, user_id) %>%
  count(step_id, name = "n_users_discovered")

step_views <- events_data %>%
  filter(action == "viewed") %>%
  distinct(step_id, user_id) %>%
  count(step_id, name = "n_users_viewed")

step_starts <- events_data %>%
  filter(action == "started_attempt") %>%
  distinct(step_id, user_id) %>%
  count(step_id, name = "n_started_attempt")

step_passed <- events_data %>%
  filter(action == "passed") %>%
  distinct(step_id, user_id) %>%
  count(step_id, name = "n_passed")
```

```r
steps_table <- steps_table %>%
  left_join(step_discovered, by = "step_id") %>%
  left_join(step_views,  by = "step_id") %>%
  left_join(step_starts, by = "step_id") %>%
  left_join(step_passed, by = "step_id")
```

```r
steps_table <- steps_table %>%
  mutate(
    n_users_viewed     = if_else(is.na(n_users_viewed), 0L, n_users_viewed),
    n_started_attempt  = if_else(is.na(n_started_attempt), 0L, n_started_attempt),
    n_passed           = if_else(is.na(n_passed), 0L, n_passed),
    start_rate   = if_else(n_users_viewed > 0,
                           n_started_attempt / n_users_viewed, NA_real_),
    pass_rate    = if_else(n_started_attempt > 0,
                           n_passed / n_started_attempt, NA_real_),
    dropoff_rate = 1 - pass_rate
  )
```

```r
library(dplyr)
library(data.table)

# Создаем events без экстремальных пользователей
events_no_extremes <- events_data %>% filter(!user_id %in% extreme_user_ids)

# Считаем метрики без экстремальных пользователей
step_views_no_ext <- events_no_extremes %>%
  filter(action == "viewed") %>%
  distinct(step_id, user_id) %>%
  count(step_id, name = "n_users_viewed_no_ext")

step_starts_no_ext <- events_no_extremes %>%
  filter(action == "started_attempt") %>%
  distinct(step_id, user_id) %>%
  count(step_id, name = "n_started_attempt_no_ext")

step_passed_no_ext <- events_no_extremes %>%
  filter(action == "passed") %>%
  distinct(step_id, user_id) %>%
  count(step_id, name = "n_passed_no_ext")

# Создаем steps_table_no_ext
steps_table_no_ext <- steps_table %>%
  left_join(step_views_no_ext, by = "step_id") %>%
  left_join(step_starts_no_ext, by = "step_id") %>%
  left_join(step_passed_no_ext, by = "step_id") %>%
  mutate(
    n_users_viewed_no_ext    = if_else(is.na(n_users_viewed_no_ext), 0L, n_users_viewed_no_ext),
    n_started_attempt_no_ext = if_else(is.na(n_started_attempt_no_ext), 0L, n_started_attempt_no_ext),
    n_passed_no_ext          = if_else(is.na(n_passed_no_ext), 0L, n_passed_no_ext),
    start_rate_no_ext   = if_else(n_users_viewed_no_ext > 0,
                                  n_started_attempt_no_ext / n_users_viewed_no_ext, NA_real_),
    pass_rate_no_ext    = if_else(n_started_attempt_no_ext > 0,
                                  n_passed_no_ext / n_started_attempt_no_ext, NA_real_),
    dropoff_rate_no_ext = 1 - pass_rate_no_ext
  )

# Метрики для расчета
step_metrics <- c("n_users_viewed", "n_started_attempt", "n_passed", "start_rate", "pass_rate", "dropoff_rate")

# Полные данные
stats_full <- rbindlist(lapply(step_metrics, function(m) {
  data.table(
    metric = m,
    mean   = round(mean(steps_table[[m]], na.rm = TRUE), 2),
    median = round(median(steps_table[[m]], na.rm = TRUE), 2),
    p90    = round(quantile(steps_table[[m]], 0.9, na.rm = TRUE), 2)
  )
}))

# Без экстремальных пользователей
stats_no_ext <- rbindlist(lapply(step_metrics, function(m) {
  data.table(
    metric = m,
    mean   = round(mean(steps_table_no_ext[[paste0(m, "_no_ext")]], na.rm = TRUE), 2),
    median = round(median(steps_table_no_ext[[paste0(m, "_no_ext")]], na.rm = TRUE), 2),
    p90    = round(quantile(steps_table_no_ext[[paste0(m, "_no_ext")]], 0.9, na.rm = TRUE), 2)
  )
}))

# Выводим таблицы
cat("=== Статистика по шагам: Полные данные ===\n")
print(stats_full)

cat("\n=== Статистика по шагам: Без экстремальных пользователей ===\n")
print(stats_no_ext)
```
